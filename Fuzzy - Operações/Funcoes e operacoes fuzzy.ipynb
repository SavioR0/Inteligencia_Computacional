## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

## Plot Gráficos

```python
def plotCharts (y, step, mini, maxi):
    x = np.array(np.arange(mini, maxi, step))
    for i, array in enumerate(y):
        plt.plot(x,array)
    
```

## Funções de Pertinência

```python
#Funções de pertinência

#Triangular
def triangle (x, a, m, b):
    vet = []
    for var in x:
        vet.append(max(min((var-a)/(m-a), (b-var)/(b-m)), 0))
    return vet

#Trapezoidal
def trapeze (x, a, m,n, b):
    vet = []
    for var in x:
        vet.append(max(min((var-a)/(m-a), 1, (b-var)/(b-n)), 0))
    return vet

#Trapezoidal
def gaussian (x, k, m):
    vet = []
    for var in x:
        vet.append(math.exp((-k*(var-m)**2)))
    return vet
```

```python
#Variações de x- passo
x = np.arange(0, 10, 0.1)

#array de funções
y1 = []

#Funções triangulares
y11 = triangle(x, 4,5,6)
y1.append(y11)
y21 = triangle(x, 3.5,5,6.5)
y1.append(y21)
y31 = triangle(x, 3,5,7)
y1.append(y31)
y41 = triangle(x, 2.5,5,7.5)
y1.append(y41)
y51 = triangle(x, 2,5,8)
y1.append(y51)
y61 = triangle(x, 1.5,5,8.5)
y1.append(y61)
y71 = triangle(x, 1,5,9)
y1.append(y71)


#Plot Funções Triangulares
plotCharts(y1, 0.1, 0, 10 )
```

# Complemento            

## Zadeh

```python
# Função de complemento Zadeh
def complementZadeh (y):
    vet = []
    for var in y:
        vet.append((1 - var))
    return vet
```

```python
c11 = []
c11.append(complementZadeh(y11))
c11.append(complementZadeh(y12))
c11.append(complementZadeh(y13))

plotCharts(c11, 0.1, 0, 10)
```

## Sugeno

```python
# Função de complemento Sugeno
def complementSugeno (y, s): # 
    vet = []
    for var in y:
        vet.append((1 - var)/(1+s*var))
    return vet
```

```python
c21 = []
# Valor de s : -1 < s < 0
c21.append(complementSugeno(y11, -0.5))
c21.append(complementSugeno(y12, -0.5))
c21.append(complementSugeno(y13, -0.5))

plotCharts(c21,  0.1, 0, 10)
```

```python
c22 = [] 
# Valor de s : s = 0
c22.append(complementSugeno(y11, 0))
c22.append(complementSugeno(y12, 0))
c22.append(complementSugeno(y13, 0))

plotCharts(c22,  0.1, 0, 10)
```

## Yager

```python
# Função de complemento Yager
def complementYager (y, w):
    vet = []
    for var in y:
        vet.append((1 - var**w)**(1/w))
    return vet
```

```python
c31 = []
#Valor de w : a < w < 0
c31.append(complementYager(y11, 0.1))
c31.append(complementYager(y12, 0.1))
c31.append(complementYager(y13, 0.1))

plotCharts(c31,  0.1, 0, 10)
```

# União

## Máximo


```python
# Função de união - Máximo
def unionMax (y1, y2):
    vet = []
    for i in range(len(y1)): 
        vet.append(max(y1[i], y2[i]))
    return vet
```

```python
u1 = []

func = triangle(x, 4.5,6,7.5)

u1.append(y41)
u1.append(func)
u1.append(unionMax(y41, func))

plotCharts(u1,  0.1, 0, 10)
```

## Soma Probabilística

```python
# Função de união - Soma Probabilística
def probabilisticSum (y1, y2):
    vet = []
    for i in range(len(y1)): 
        vet.append(y1[i] + y2[i]- y1[i]*y2[i])
    return vet
```

```python
u2 = []
#func = triangle(x, 4.5,6,7.5)
u2.append(y41)
u2.append(func)
u2.append(probabilisticSum(y41, func))

plotCharts(u2,  0.1, 0, 10)
```

## Soma Limitada

```python
# Função de união - Soma Limitada
def limitedSum (y1, y2):
    vet = []
    for i in range(len(y1)): 
        vet.append(min(1, y1[i]+y2[i]))
    return vet
```

```python
u3 = []
#func = triangle(x, 4.5,6,7.5)
u3.append(y41)
u3.append(func)
u3.append(limitedSum(y41, func))

plotCharts(u3,  0.1, 0, 10)
```

## Soma Drástica

```python
# Função de união - Soma Drástica
def drasticSum (y1, y2):
    vet = []
    for i in range(len(y1)): 
        if (y1[i] == 0):
            vet.append(y2[i])
            continue
        elif (y2[i] == 0):
            vet.append(y2[i])
            continue
        vet.append(1)
    return vet
```

```python
u3 = []
#func = triangle(x, 4.5,6,7.5)
u3.append(y41)
u3.append(func)
u3.append(drasticSum(y41, func))

plotCharts(u3,  0.1, 0, 10)
```

## Mínimo

```python
# Função de Interseção - Mínimo
def interMin (y1, y2):
    vet = []
    for i in range(len(y1)): 
        vet.append(min(y1[i], y2[i]))
    return vet
```

## Produto

```python
# Função de insterseção - Produto
def interProd (y1, y2):
    vet = []
    for i in range(len(y1)): 
        vet.append(y1[i]* y2[i])
    return vet
```

```python
i2 = []

#func = triangle(x, 4.5,6,7.5)

i2.append(y41)
i2.append(func)
i2.append(interProd(y41, func))

plotCharts(i2,  0.1, 0, 10)
```

## Produto Limitado

```python
# Função de insterseção - Produto Limitado 
def interProdLimited (y1, y2):
    vet = []
    for i in range(len(y1)): 
        vet.append(max(0,y1[i]+y2[i]-1))
    return vet
```

```python
i3 = []

#func = triangle(x, 4.5,6,7.5)

i3.append(y41)
i3.append(func)
i3.append(interProdLimited(y41, func))

plotCharts(i3,  0.1, 0, 10)
```

## Produto Drástico

```python
# Função de insterseção - Produto Drástico 
def interProdDrastic (y1, y2):
    vet = []
    for i in range(len(y1)):
        if y2[i] == 1:
            vet.append(y1[i])
            continue
        elif y1[i] == 1:
            vet.append(y2[i])
            continue 
        vet.append(0)
    
    return vet
```
