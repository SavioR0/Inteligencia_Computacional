```python
from random import *
import numpy as np
```

```python
class MLP:
    def __init__ (self, learningRate, maxEpoch, tol, bias, inputs, output):
        self.learningRate = learningRate
        self.epoch = maxEpoch
        #O Bias de todos os neurônios é igual a -1
        self.bias = bias
        self.tol = tol
        
        self.inputs = inputs
        self.output = output
        
        self.errorEpoch = tol + 1;
        self.nEpoch = 0
        
        self.weights = {"w31": uniform(-0.5, 0.5),
           "w3O1": uniform(-0.5, 0.5),
           "w41": uniform(-0.5, 0.5),
           "w4O2": uniform(-0.5, 0.5),
           "w32": uniform(-0.5, 0.5),
           "w42": uniform(-0.5, 0.5),
           "w53": uniform(-0.5, 0.5),
           "w54": uniform(-0.5, 0.5),
           "w5O3": uniform(-0.5, 0.5),
                        
          }
    
    def activation(self,inputs):
        if inputs >= 0.0 : 
            return 1
        return 0
        
    def fit(self):
        evec=[]
        while(self.errorEpoch>tol and self.nEpoch < maxEpoch):
            print(f"------------ EPOCA {self.nEpoch}------------")
            error = 0
            for i in range(len(self.inputs)):
                # nm Cáuculo da Camada intermediária no neurônio m
                n3 = self.inputs[i][0]*self.weights["w" + str(3)+str(1)] + self.inputs[i][1]*self.weights["w" + str(3)+str(2)]+ self.bias*self.weights["w"+str(3)+"O"+str(1)]
                n4 = self.inputs[i][0]*self.weights["w" + str(4)+str(1)] + self.inputs[i][1]*self.weights["w" + str(4)+str(2)]+ self.bias*self.weights["w"+str(4)+"O"+str(2)]
                # in é a saida do neurônio n
                i3 = self.activation(n3)
                i4 = self.activation(n4)
                # Neurônio Camada de saída
                n5 = i3*self.weights["w" + str(5)+str(3)] + i4*self.weights["w" + str(5)+str(4)] + bias*self.bias*self.weights["w"+str(5)+"O"+str(3)]
                # Saída do neurônio Camada de saída
                i5 = self.activation(n5)
                print("\t Saída : " + str(i5), end = '')
                
                # Calculo do erro
                e5 = self.output[i] - i5
                
                # Calculos das variações 
                # Influencia do neuronio 5
                delta5 = e5 * ((1/np.cosh(n5).item())**2)
                dw5O3 = learningRate*delta5*bias
                dw53 =  learningRate*delta5*i3
                dw54 =  learningRate*delta5*i4
                
                # Influencia do neuronio 3 
                delta3 = (delta5*self.weights["w"+ str(5)+ str(3)])*(1/np.cosh(n3))**2
                dw3O1 = learningRate*delta3*bias
                dw31 =  learningRate*delta3*self.inputs[i][0]
                dw32 =  learningRate*delta3*self.inputs[i][1]
                
                # Influencia do neuronio 4 
                delta4 = (delta5*self.weights["w"+str(5)+str(4)])*(1/np.cosh(n4))**2
                dw4O2 = learningRate*delta4*bias
                dw41 =  learningRate*delta4*self.inputs[i][0]
                dw42 =  learningRate*delta4*self.inputs[i][1]
                
                # Atualizações do pesos
                self.weights["w5O3"] = self.weights["w5O3"] + dw5O3
                self.weights["w53"] = self.weights["w53"] + dw53
                self.weights["w54"] = self.weights["w54"] + dw54
                self.weights["w3O1"] = self.weights["w3O1"] + dw3O1
                self.weights["w31"] = self.weights["w31"] + dw31
                self.weights["w32"] = self.weights["w32"] + dw32
                self.weights["w4O2"] = self.weights["w4O2"] + dw4O2
                self.weights["w41"] = self.weights["w41"] + dw41
                self.weights["w42"] = self.weights["w42"] + dw42
                
                error = error + e5**2
                print(f"\t Erro : {error}")
            self.errorEpoch = error/len(self.inputs)
            self.nEpoch = self.nEpoch + 1;
            print("--------------------------------")
            #evec[self.nEpoch] = self.errorEpoch
        print(f"Pesos escolhidos : {self.weights}")
                
                
```

```python
maxEpoch = 100
learningRate = 0.2
tol =0.2
bias = -1

# XOR
inputs = [[0,0],
          [0,1],
          [1,0],
          [1,1]] 

output = [0,1,1,0]


mlp = MLP(learningRate, maxEpoch, tol, bias, inputs, output)
mlp.fit()
```
